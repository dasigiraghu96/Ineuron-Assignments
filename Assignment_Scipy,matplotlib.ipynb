**We have the min and max temperatures in a city In India for each months of the year.
We would like to find a function to describe this and show it graphically, the dataset
given below.
Task:
1.fitting it to the periodic function
2.plot the fit
Data
Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25
Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18**

```
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
x = range(0,12)
T_max =np.asarray([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
T_min=np.asarray([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
plt.figure(figsize=(6, 4))
np.random.seed(0)

from scipy import optimize
def test_func(x, a, b,c,d):
    return a*np.sin(b * x+d)+c
def test_func1(x, a, b,c,d):
    return a*np.sin(b * x+d)+c

fig=plt.figure(figsize=(8,5))
params1, params_covariance1 = optimize.curve_fit(test_func, x,T_max)
plt.scatter(x,T_max,label='Max. Temperature')
plt.plot(x, test_func(x, params1[0], params1[1],params1[2],params1[3]),label='Max Temp function')

params2, params_covariance2 = optimize.curve_fit(test_func1, x,T_min)
plt.scatter(x,T_min,label='Min. Temperature')
plt.plot(x, test_func1(x, params2[0], params2[1],params2[2],params2[3]),label='Min Temp function')
plt.legend(loc='best')
plt.xlabel('Month')
plt.ylabel('Temperature deg C')
```
